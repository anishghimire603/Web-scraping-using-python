```python
import requests
import urllib.request
from bs4 import BeautifulSoup
```

```python
url = 'https://yts.mx/'
response =requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text,'html.parser')
print(soup.prettify()) //indentation 
```

```python
def download_pic(photo_url):
    name = photo_url.split('/')
    name = name[-2] + name[-1]
   # print(name)
    urllib.request.urlretrieve(photo_url, name)  
  
    photos = soup.findAll('div',{'class':'browse-movie-wrap col-xs-10 col-sm-5'})
for photo in photos:
    figure = photo.find('figure')
    print(figure)
    photo_url = 'https://img.yts.mx/' + figure.img['src']
    print(photo_url)
    download_pic(photo_url)
   
```
